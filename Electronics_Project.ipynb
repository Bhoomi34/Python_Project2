```python
#bar plot
import matplotlib.pyplot as plt

#Assuming 'category' is a list of categories
category_counts=dataset['category'].value_counts()
#type of plot
plt.bar(category_counts.index,category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of items in Each category')
plt.xticks(rotation=45) #rotate x-axis labels for better readability #optional
plt.show()
```

```python
#histogram is a distribution of data(frequency)
plt.hist(dataset['rating'],bins=10) #bin is an attribute which decides the width of the graph #more the number lesser the width
plt.xlabel('Rating') #value
plt.ylabel('Frequency') #frequency
plt.title('Distribution of Ratings')
plt.show()
```

```python
model_attr_counts=dataset['model_attr'].value_counts()
#input parameter:model_attr_counts
plt.pie(model_attr_counts,labels=model_attr_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Model Attributes')
plt.show
```
